```python
import pandas as pd
```

```python
df=pd.read_csv('corona_dat.csv')
```

```python
df.head()
```

```python
df2=df[['date','China','Italy','Spain','France','United Kingdom','Brazil','US']]
```

```python
df2.set_index('date',inplace=True)
```

```python
df2.head()
```

```python
df_total=df2.cumsum(axis=0) #cumulative sum(cumsum)
```

```python
import bar_chart_race as bcr
%matplotlib inline
```

```python
bcr.bar_chart_race(df_total,filename='covid19.mp4', figsize=(8,5),title='Gilly')
```
