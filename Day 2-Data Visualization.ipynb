### Feature Engineering

```python
#feature/property/predictor/attributes also columns(not preferred) engineering
# this is generating new columns for the dataset
```

```python
df['day']=(df['Date']).dt.day
df['Month']=(df['Date']).dt.month
df['Year']=(df['Date']).dt.year
```

```python
df.head()
```

```python
df['month_name']=df['Date'].dt.month_name()
df['weekday']=df['Date'].dt.day_name()
```

```python
df.head()
```

```python
sns.countplot(x=df['weekday'])
plt.xticks(rotation=90);
```

```python
df['Time']=pd.to_datetime(df.Time)
```

```python
df['Hour']=(df['Time']).dt.hour
```

```python
df.head()
```

```python
sns.lineplot(x='Hour',y='Quantity',data=df)
```

```python
sns.pairplot(df)
```

```python
sns.relplot(x='Hour',y='Quantity',col='Month',row='Branch', kind='line',
          hue='Gender',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df["gross income"],y=df["Product line"])
plt.xticks(rotation=90);
```
