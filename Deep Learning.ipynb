```python
df=pd.read_csv('kcdata.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('price',axis=1)
y=df[['price']]
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1))#Regression
model.compile(loss='mse',optimizer='adam') # regression,metrics=mse
#Early stopping
```

```python
history=model.fit(x_train,y_train, 
                  validation_data=(x_test,y_test),epochs=150, batch_size=128, verbose=1)
```

```python
model.summary()
```

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, pred)
```

```python
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist
(train_images,train_labels),(test_images,test_labels)=mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
train_images[59999].shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D,MaxPooling2D,Flatten, InputLayer, Reshape
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(28,28)))
model.add(Reshape(target_shape=(28,28,1)))
model.add(Conv2D(filters=12, kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(10)) # Classification 
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),optimizer='adam',metrics='accuracy')
```

```python
model.fit(train_images,train_labels,validation_split=0.10,epochs=10,verbose=1)
```

```python
model.save('mnist.db5')
```
